```python
i = 0
while(i<150):
    driver.execute_script("window.scrollBy(0,200)")
    i+=1
    time.sleep(3)
    
    
Nike_Shoe_price = []    
try:
    price= driver.find_elements(By.XPATH,'//div[@class="product-card__animation_wrapper"]')[:200]
    for pr in price:
        Nike_Shoe_price.append(pr.text)
except NoSuchElementException:
     Nike_Shoe_price.append('_')
        
    
Nike_Shoe_Name = []
try:
    Name= driver.find_elements(By.XPATH,'//div[@class="product-card__title"]')[:200]
    for na in Name:
        Nike_Shoe_Name.append(na.text)
except NoSuchElementException:
    Nike_Shoe_Name.append('_')
    
    
Nike_Shoe_Number_of_colour = []
try:
    Number_of_colour= driver.find_elements(By.XPATH,'//div[@class="product-card__count-item"]')[:200]
    for co in Number_of_colour:
        Nike_Shoe_Number_of_colour.append(co.text)
except NoSuchElementException:
    Nike_Shoe_Number_of_colour.append('_')
    

opening_url = []    
url = driver.find_elements(By.XPATH,'//a[@class="product-card__link-overlay"]')[:200]
for i in url:
    opening_url.append(i.get_attribute('href')) 
```
